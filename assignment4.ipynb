```python
x = input("Enter number of inputs :")
x = int(x)
database_list = []
for i in range(1,x+1):
    degree = {}
    database = {}
    lst = []
    name = input("Enter name :")
    age = input("Enter age :")
    age = int(age)
    work = input("Enter work :")
    skill_input = input("Enter number of skills :")
    skill_input = int(skill_input)
    for j in range(1,skill_input + 1):
        skill = input("Enter skill :")
        lst.append(skill)
    degree_title = input("Enter title :")
    degree_major = input("Enter major :")
    degree_completionDate = input("Enter completiondate (month year ) :")
    salary = input("Enter salary :")
    salary = int(salary)
    database["name"] = name
    database["age"] = age
    database["work"] = work
    database["skills"] = lst
    degree["title"] = degree_title
    degree["major"] = degree_major
    degree["completionDate"] = degree_completionDate
    database["degree"] = degree
    database["salary"]  = salary
    database_list.append(database)
    
print (database_list)
    
```
